## Puzzle 3: The `p` situation

```python
def clean_item_price(item_price: str):
    if item_price.startswith("£"):
        cleaned_price = item_price[1:].strip()
        return cleaned_price
    elif item_price.endswith("p"):
        cleaned_price = item_price[0:-1].strip()
        cleaned_price = '0.' + cleaned_price
        return cleaned_price

print(clean_item_price("£9.99"))
print(clean_item_price("99p"))
```

## Puzzle 4: The `each` situation

```python
def clean_item_price(item_price: str):
    cleaned_price = item_price #ensure cleaned_price processes each condition in sequence
    if item_price.endswith("p"):
        cleaned_price = item_price.replace('p', '').strip()
        cleaned_price = '0.' + cleaned_price
        
    if 'each est.' in cleaned_price:
        cleaned_price = cleaned_price.replace('each est.', '')
        
    if '£' in cleaned_price:
        cleaned_price = cleaned_price.replace('£', '').strip()
      
    return float(cleaned_price)

clean_item_price("£9.99 each est.")
```
